```python
master_dataset = cw.load_master_data()
```

```python
PV_control_neuron_dict = {};

for PV_control_neuron in master_dataset['names']:
    if 'PV1' in PV_control_neuron:
        #print PV_control_neuron
        #print master_dataset['names'].index(PV_control_neuron)
        master_index_value = master_dataset['master_index'][master_dataset['names'].index(PV_control_neuron)];
        #print master_dataset['isNeuronList'][master_index_value]
        
        PV_control_neuron_dict[PV_control_neuron] = master_dataset['isNeuronList'][master_index_value];
        
    if 'PV2' in PV_control_neuron:
        #print PV_control_neuron
        #print PV_controls_dataset['names'].index(PV_control_neuron)
        master_index_value = master_dataset['master_index'][master_dataset['names'].index(PV_control_neuron)];
        PV_control_neuron_dict[PV_control_neuron] = master_dataset['isNeuronList'][master_index_value];
```

```python
import os;

PV_control_neuron_results = {};

for PV_control_neuron in PV_control_neuron_dict.keys():
    print "Calculating for " + PV_control_neuron;
    if sum(PV_control_neuron_dict[PV_control_neuron]) == 0:
        print "No neurons for " + PV_control_neuron + " dataset!";
    else:
        data = cw.load_data_from_MATLAB('/scratch/users/ihong2@jhu.edu/Data/' + PV_control_neuron + '.mat');
        orientation_dict, averaged_firing_dict, neuronNumber, trial = cw.calculate_activity(data);
        isNeuronList = PV_control_neuron_dict[PV_control_neuron]
        L_ori, L_dir = cw.calculate_normalized_orientation_and_direction(orientation_dict, averaged_firing_dict, neuronNumber, trial, plot_neuron_list = isNeuronList);
        L_ori_plot = cw.barplot_L_ori(L_ori, filename = PV_control_neuron + '_L_ori.html');
        CIRVAR, DIRVAR = cw.calculate_CIRVAR_and_DIRVAR(L_ori, L_dir);
        CIRVAR_plot = cw.barplot_CIRVAR(CIRVAR, filename = PV_control_neuron + '_CIRVAR.html');
        final_subplots = cw.subplots(orientation_dict, averaged_firing_dict, neuronNumber, trial, plot_neuron_list = isNeuronList, row = 10, col = 10, filename = PV_control_neuron + '_subplots.html');

        for neuron in PV_control_neuron_dict[PV_control_neuron]:
            orientation_plot = cw.orientation_polar_plot(orientation_dict, averaged_firing_dict, neuronNumber, trial, plot_neuron_list=[neuron], isScaled=True, save_to_plotly=True, filename= PV_control_neuron + '_' + str(neuron) + '_scaled_orientation_polar_plot.html');
            orientation_name = 'orientation_neuron_' + str(neuron);

        results = {};
        results['isNeuronList'] = isNeuronList;
        results['L_ori'] = L_ori;
        results['L_dir'] = L_dir;
        results['CIRVAR'] = CIRVAR;
        results['DIRVAR'] = DIRVAR;
        results['orientation_dict'] = orientation_dict;
        results['averaged_firing_dict'] = averaged_firing_dict;
        results['neuronNumber'] = neuronNumber;
        results['trial'] = trial;

        print "Dumping file as a pickle in pickle folder..."
        if not os.path.exists('pickle'):
            os.makedirs('pickle')

        if not os.path.exists('pickle/' + PV_control_neuron):
            os.makedirs('pickle/' + PV_control_neuron)

        with open('pickle/' + PV_control_neuron + '/' + PV_control_neuron + '_results.pickle', 'wb') as handle:
            pickle.dump(results, handle, protocol=pickle.HIGHEST_PROTOCOL)

        PV_control_neuron_results[PV_control_neuron] = results;
    
```

```python
#  Dump saved results in pickle so we can access it again later.

with open('pickle/PV_control_results.pickle', 'wb') as handle:
    pickle.dump(PV_control_neuron_results, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
