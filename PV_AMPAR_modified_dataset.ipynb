```python
import calcium_workflow as cw
reload(cw)
```

```python
master_dataset = cw.load_master_data()
```

```python
PV_AMPAR_neuron_dict = {};

for PV_AMPAR_neuron in master_dataset['names']:
    if 'PV8' in PV_AMPAR_neuron:
        master_index_value = master_dataset['master_index'][master_dataset['names'].index(PV_AMPAR_neuron)];        
        PV_AMPAR_neuron_dict[PV_AMPAR_neuron] = master_dataset['isNeuronList'][master_index_value];
        
    if 'PV9' in PV_AMPAR_neuron:
        master_index_value = master_dataset['master_index'][master_dataset['names'].index(PV_AMPAR_neuron)];
        PV_AMPAR_neuron_dict[PV_AMPAR_neuron] = master_dataset['isNeuronList'][master_index_value];
```

```python
PV_AMPAR_results = cw.run_pipeline(PV_AMPAR_neuron_dict)
```

```python
#  Dump saved results in pickle so we can access it again later.

import pickle
with open('pickle/PV_AMPAR_results.pickle', 'wb') as handle:
    pickle.dump(PV_AMPAR_results, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
